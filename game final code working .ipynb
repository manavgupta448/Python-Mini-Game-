```python
import pygame  #pygame -used to create the game window 
import random  #random -to generate random shadow positions 

pygame.init()  #pygame.init() - to initialize pygame so that we can use its features 



font = pygame.font.SysFont(None, 30)  # Default font and size
start_time = pygame.time.get_ticks()  # Record start time in milliseconds



# Game window dimensions
width, height = 500, 400  #it will define the size of the game window 
screen = pygame.display.set_mode((width, height))  #it will create the game window 
pygame.display.set_caption("Shadow Dodge")  #it will set the window title 

# Colors
#rgb colors content in the screen.all colors are made of white black blue with different proportions.
white = (255, 255, 255)
black = (0, 0, 0)
blue = (0, 0, 255)

# Player properties
player_size = 35  #the player is a blue square with size 35x35
player_x = width // 2 - player_size // 2  #to start in the middle
player_y = height - 50  #near the bottom
player_speed = 7  # it will control how fast it will move

# Shadow (enemy) properties
shadow_width = 60  
shadow_height = 20  
shadow_speed = 4  # Starting speed, will increase over time
shadows = []  
#shadows are rectangles with dimensions 60x20 and speed of moving as 4

# Game variables
clock = pygame.time.Clock()  #(it controls the speed of the game ,that is frame rate)
running = True  #it keeps the game running until the player is hit 
score = 0  #it is used to track the player progress

# Function to create shadows
def create_shadow():
    x_pos = random.randint(0, width - shadow_width)   #this command will choose a random position for the shadow 
    shadows.append([x_pos, 0])  #the shadow will start at y=0(that is at the top of the screen )
#each shadow will be stored as [x,y] in the shadows list .

# Main game loop
#this loop will run continuously until the game ends 



# Calculate and display time survived
elapsed_time = (pygame.time.get_ticks() - start_time) // 1000  # seconds
timer_text = font.render(f"Time: {elapsed_time}s", True, black)
screen.blit(timer_text, (10, 10))  # Draw it at top-left

  #  background color
yellow = (255, 255, 0)  # RGB for yellow
background_color = yellow


while running:
    clock.tick(30)  #it will control the speed of the game that is it will limit the game to 30 frames per second
    screen.fill(background_color)  #it will clear the trails that is the path of shadow in every frame 

    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
#it is used for closing the game window once the user is done 
#when the user will click the cross button ,running=false , will close the game window 

    # Player movement
    keys = pygame.key.get_pressed()  #it will determine which keys are being pressed #keys is a list where each key's state pressed or not pressed is stored
    if keys[pygame.K_LEFT] and player_x > 0:  # keys[pygame.K_LEFT]:it will check if left arrow key is pressed  and player_x>0:it will check if the player doesn't move off the edge
        player_x -= player_speed  #it will move the player left by player_speed pixels
    if keys[pygame.K_RIGHT] and player_x < width - player_size: #keys[pygame.K_RIGHT]::it will check if right arrow key is being pressed ,player_x < WIDTH - player_size will check if the player doesn't move off the right edge 
        player_x += player_speed  # it will move the player right by player_speed pixels

    # Shadow spawning (randomized)
    if random.randint(1, 30) == 1:  
        create_shadow()

#random.randint(1, 30) == 1 :generates a random number between 1 and 30,== 1: The condition is true only if the random number is exactly 1.
#when true, a new shadow is created by calling create_shadow()
#so instead of creating shadows at a fixed time interval ,this method introduces randomness:in every frame,the game checks if a shadow should be created
#there is a 1 in 30 probability that a new shadow will appear
#it is used because if shadows were created at fixed interval ,the game would feel predictable .now sometimes many shadows would appear quickly ,or less

    # Move and draw shadows
    for shadow in shadows:
        shadow[1] += shadow_speed  #each shadow moves down with shadow_speed pixels per frame ,
        pygame.draw.rect(screen, black, (shadow[0], shadow[1], shadow_width, shadow_height))  #pygame.draw.rect :: draws each shadow on the screen 

        # Collision detection
        if (shadow[1] + shadow_height >= player_y and  
            shadow[0] < player_x + player_size and  
            shadow[0] + shadow_width > player_x):
            running = False  # Game Over

#shadow[1] + shadow_height >= player_y: the shadow reaches the player's height.
#shadow[0] < player_x + player_size: the shadow's left side is inside the player.
#shadow[0] + shadow_width > player_x: the shadow's right side is inside the player.

    # Remove off-screen shadows
    shadows = [shadow for shadow in shadows if shadow[1] < height]  
#it will filter the shadows list  to remove any  shadow that has moved off screen ,once a shadow's position is longer than height ,it will no longer be visible on screen

    # Increase speed after every 10 points
    score += 1  # Increase score as player survives  
    if score % 300 == 0:  # Every 300 frames (~10 seconds at 30 FPS)
        shadow_speed += 1  

    # Draw player
    pygame.draw.rect(screen, blue, (player_x, player_y, player_size, player_size))  
#it will draw the blue player  square  at the updated position

    # Display the game
    pygame.display.update()
#it will refresh the screen every frame to show updated positions

pygame.quit()
#it will close the game properly when the loop ends
```
